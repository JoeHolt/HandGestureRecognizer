```python
def cvt_bw(img):
    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    return cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY)[1]
```

```python
def create_mask(fg):
    fg_bw = cvt_bw(fg)
    mask_idx = (fg_bw == 255)
    return mask_idx
```

```python
def apply_mask(fg, mask):
    new_img = copy(fg)
    new_img[mask] = 0
    return new_img
```
